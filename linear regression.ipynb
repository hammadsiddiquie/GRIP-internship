### Step-1: Importing Data using Pandas library

```python
import pandas as pd
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
print(df.head())
```

### Step-2: Plotting the Dataframe

```python
import matplotlib.pyplot as plt 
df.plot(kind = 'scatter',x = 'Hours', y = 'Scores')
```

### Step-3: Install sci-kit Learn library

```python
!pip install scikit-learn
```

### Step-4: Splitting Data into test and train sets

```python
x = df[['Hours']]
y = df['Scores']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

### Step-5: Training Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train, y_train)
print(model)
```

```python
plt.scatter(x_train, y_train)
plt.xlabel("Hours")
plt.ylabel("salary")

plt.plot(x_train,model.predict(x_train))
plt.show()
```

### Step-6: Calculating the model's score based on test and train datasets

```python
score = model.score(x_test, y_test)
print("Test scores is:",score)
```

```python
score = model.score(x_train, y_train)
print("Training scores is:",score)
```

```python
y_predicted = model.predict(x_test)
print("Predicted scores are", y_predicted)
```

## what will be the predicted score if student studies 9.25 hrs/day?

```python
y_predicted = model.predict([[9.25]])
print("predicted score, if student studies 9.25 hrs/day is:", y_predicted)
```

```python
df = pd.DataFrame({'Actual Score': y_test, 'Predicted Score': y_predicted})  
df 
```

### Step-6: Calculating Mean square error

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_predicted)) 
```
